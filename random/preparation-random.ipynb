# Preperation Random Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.DataFrame(np.random.randint(0, 1000, size=(100, 4)), columns=list(['x1', 'x2', 'y1', 'y2']))
```

```python
col_name = df.columns.tolist()
col_name.insert(0,'time')
col_name.insert(1, 'group')
df = df.reindex(columns=col_name)
```

```python
for i in range(len(df.group)):
    df.group[i] = random.randint(1,5)
```

```python
df.time = range(100)
```

```python
df.head()
```

# Distribution of links

```python
plt.plot(df.x1, df.x2)
plt.show()
```

# Statistic of groups

```python
group_index = df.groupby(['group'],as_index=False)['group'].agg({'cnt':'count'})
```

```python
plt.bar(group_index.group, group_index.cnt)
plt.show()
```

```python
df.to_csv("random-data.csv", sep=',', encoding='utf-8', index=False)
```
