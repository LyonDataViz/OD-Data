# Taxi OD data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("TaxiOD.csv") 
```

# Example of dataset

```python
df.head()
```

# Plot trajectories

```python
plt.figure('asdf')
aa = plt.gca()
aa.plot([df.oLon, df.dLon], [df.oLat, df.dLat], color='r', linewidth=0.1, alpha=0.5)
plt.show(aa)
```

# Distribution of average speed for all trajectories

```python
df.aSpeed.hist(bins=20)
```

```python
df.oTslot.hist(bins=24)
```
